# Naive Bayesian Classifier DIY

Let us have a dataset of instances that are labeled as objects of different classes(*c* - state or class, number of classes n=2 in my training dataset). What we want to be able to do: to predict the class of a new instance.

Naive Bayesian Classifier works fast and cool on datasets with independent features. The idea used is that the conditional probability of vector x(features), assuming *c* is the product of conditional probabilities of every feature, assuming this class. And then use this formula from the basics of Theory of Probability.



### Importing dataset

```python
# Install a pip package in the current Jupyter kernel
import sys
!{sys.executable} -m pip install pandas --user

import pandas as pd

X=pd.read_csv('data_pos.txt', sep="\t", header=None)
X.head()
Z=pd.read_csv('data_neg.txt', sep="\t", header=None)
```

So we have our label at the first column: 1.0 states for true or positive in this **data_pos.txt** and -1.0 states for negative in **data_neg.txt**.

Lets plot mapped instances, since we only have 2 features - second and third columns.


```python
import matplotlib.pyplot as plt

X.columns = ['class', 'x1', 'x2']
X['x1']=X['x1'].astype(int)
X['x2']=X['x2'].astype(int)
Z.columns=['class', 'x1', 'x2']
Z['x1']=Z['x1'].astype(int)
Z['x2']=Z['x2'].astype(int)
plt.xlabel("x1")
plt.ylabel("x2")
count_red=len(X)
plt.plot(X['x1'], X['x2'], 'ro')
plt.axis([0, 10, 0, 20])
count_green=len(Z)
plt.plot(Z['x1'], Z['x2'], 'go')
```

```python
#counting dots for frequencies
set_x1={}
set_x2={}
set_z1={}
set_z2={}
for x in X['x1'].values:
    if x in set_x1:
        set_x1[x]+=1
    else:
        set_x1[x]=1
for x in X['x2'].values:
    if x in set_x2:
        set_x2[x]+=1
    else:
        set_x2[x]=1
        
for x in Z['x1'].values:
    if x in set_z1:
        set_z1[x]+=1
    else:
        set_z1[x]=1
for x in Z['x2'].values:
    if x in set_z2:
        set_z2[x]+=1
    else:
        set_z2[x]=1
```

```python
#counting frequencies
variety_x1 = {**set_x1, **set_z1};
variety_x2 = {**set_x2, **set_z2};

m_x1, m_x2, n = len(variety_x1), len(variety_x2), 2; # since we only have two classes

# here comes the likelihood from formula`s top
# p_x1[0] stands for p(x1|class =1) and p_x1[1] stands for p(x1|class=2)
p_x1 = [{},{}];
p_x2 = [{},{}];

for x in variety_x1.keys():
    if x in set_x1:
        p_x1[0][x]=set_x1[x];
    else:
            p_x1[0][x]=0;
    if x in set_z1:
        p_x1[1][x]=set_z1[x];
    else:
            p_x1[1][x]=0;

for x in variety_x2.keys():
    if x in set_x2:
        p_x2[0][x]=set_x2[x];
    else:
            p_x2[0][x]=0;
    if x in set_z2:
        p_x2[1][x]=set_z2[x];
    else:
            p_x2[1][x]=0;
```

So what we have now is almost the **likelihood** for each feature and number (it's not divided by count_red or green yet!)
Since features are independent for Naive Bayesian to work cool to get the *p(x|class=1)* for the new instance x with features x1 and x2 we only need to multiply *p(x1|class=1)* and *p(x2|class=1)*.

So now we can use the first formula to test classifier on test data.
Or we can map the existing data like decide which class rules the region on the plot. We'll o

```python
#classes primary p
p_class1 = count_red/(count_red + count_green)
p_class2 = count_green/(count_red + count_green)

def bayesian_classifier(x, y):
        p_instance = (p_x1[0][x]*p_x2[0][y])/(count_red*count_red)+(p_x1[1][x]*p_x2[1][y])/(count_green*count_green)
    
        #could be a problem if there were no instances with such x1 and x2 in dataset
        #then change p_instance, approximate it with neighbours
        
        if p_instance==0:
        #here i blindly believe it has neighbour somewhere close, otherwise BC cannot classify in a good way
            for i in range(y-1,y+1):
                for j in range(x-1,x+1):
                    if i in p_x1[0]:
                        if j in p_x2[0]:
                            p_instance = (p_x1[0][i]*p_x2[0][j])/(count_red*count_red)+(p_x1[1][i]*p_x2[1][j])/(count_green*count_green)
                    else:
                        if i in p_x1[1]:
                            if j in p_x2[1]:
                                p_instance = (p_x1[0][i]*p_x2[0][j])/(count_red*count_red)+(p_x1[1][i]*p_x2[1][j])/(count_green*count_green)
        if p_instance==0:
            return 1
        likel_class1 = p_x1[0][x]/count_red * p_x2[0][y]/count_red * p_class1 /p_instance
    
        likel_class2 = p_x1[1][x]/count_green * p_x2[1][y]/count_green * p_class2/p_instance
        
        if likel_class1 > likel_class2:
            #ax.add_artist(plt.Circle((x, y), 0.2, color='r'))
            #plt.gcf().gca().add_artist(plt.Circle((x, y), 0.2, color='r'))
            ax.add_patch(plt.Circle((x, y), 0.2, color='r', alpha=0.5))
        else: 
            #ax.add_artist(plt.Circle((x, y), 0.2, color='g'))
            #plt.gcf().gca().add_artist(plt.Circle((x, y), 0.2, color='g'))
            ax.add_patch(plt.Circle((x, y), 0.2, color='g', alpha=0.5))
        return 0
```

No let's classify each center(everywhere where our points are in this dataset) using the classifier we wrote:)

```python
import matplotlib.patches as ptch
fig, ax = plt.subplots()


for x in variety_x1.keys():
    for y in variety_x2.keys():
       bayesian_classifier(x, y);

##test on point (2,9) or any othe points
#bayesian_classifier(2,9)

ax.set_aspect('equal', adjustable='datalim')
ax.plot()   #Causes an autoscale update.
plt.show()
```
